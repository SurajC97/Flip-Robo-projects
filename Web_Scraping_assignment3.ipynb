## Task 2

1. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Creating list to store the url of the 2nd and 3rd page.

pages=[]

page= driver.find_elements_by_xpath("//li[@class='a-normal']//a")
for i in page:
    pages.append(i.get_attribute('href'))
```

```python
# Adding the home page url, so that our program can scrape data from the 3rd page and then go to the home page.

pages.append('https://www.amazon.in/')
pages
```

```python
# creating lists to store data

brand=[]
product=[]
rate=[]
no_of_rate=[]
cost=[]
re=[]
delivery=[]
availability=[]
other=[]
urls=[]


# Scraping all the details for each product.

for i in pages:
    for j in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
        urls.append(j.get_attribute('href'))
    for k in urls:
        driver.get(k)
        time.sleep(2)
        try:
            b= driver.find_element_by_xpath("//td[@class='a-span9']//span[1]")
            brand.append(b.text)
        except NoSuchElementException:
            brand.append('-')

        try:
            p= driver.find_element_by_id("productTitle")
            product.append(p.text)
        except NoSuchElementException:
            product.append('-')

        try:
            r= driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
            rate.append(r.text)
        except NoSuchElementException:
            rate.append('-')

        try:
            n= driver.find_element_by_xpath("//span[@class='a-size-base a-color-secondary']")
            no_of_rate.append(n.text)
        except NoSuchElementException:
                no_of_rate.append('-')

        try:
            c= driver.find_element_by_xpath("//td[@class='a-span12']")
            cost.append(c.text)
        except NoSuchElementException:
            cost.append('-')

        try:
            e= driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
            re.append(e.text)
        except NoSuchElementException:
            re.append('-')

        try:
            d= driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']//b")
            delivery.append(d.text)
        except NoSuchElementException:
            delivery.append('-')

        try:
            a= driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")
            availability.append(a.text)
        except NoSuchElementException:
            availability.append('-')

        try:
            o= driver.find_element_by_id("feature-bullets")
            other.append(o.text.replace('\n',''))
        except NoSuchElementException:
            other.append('-')

    driver.find_element_by_id('nav-search-submit-button').click()   # clicking the search button to get to the 1st page.
    time.sleep(2)
    driver.get(i)    # For opening the 2nd and 3rd page.
```

```python
# Creating dataframe

x= {'Brand':brand,'Product name': product,'Rating': rate, 'Number of ratings': no_of_rate, 'Price': cost, 'Return/Exchange': re,
    'Expected Delivery': delivery, 'Availability': availability, 'Other details': other, 'Product URL': urls}
df1= pd.DataFrame(x)
df1
```

```python
# Accessing the search bar of new page and entering the next keyword.

search_bar1= driver.find_element_by_xpath("//input[@class='og3lId']")
search_bar1.clear()
search_bar1.send_keys('cars')
time.sleep(2)


# Accessing the new search button and clicking in it.

search_btn1= driver.find_element_by_xpath("//div[@class='XZ5MVe']")
search_btn1.click()
time.sleep(3)


# Scrolling the page for the data to appear.

driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
time.sleep(8)


# Scraping the images source.

cars=[]
for i in driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']"):
    cars.append(i.get_attribute('src'))
time.sleep(3)
    
# Keeping top 100 images.

cars= cars[0:100]


# Finding and clearing the search bar then entering the next keyword.

search_bar1= driver.find_element_by_xpath("//input[@class='og3lId']")
search_bar1.clear()
search_bar1.send_keys('Machine learning')
time.sleep(2)


# Accessing and clicking in search button.

search_btn1= driver.find_element_by_xpath("//div[@class='XZ5MVe']")
search_btn1.click()
time.sleep(3)


# Scrolling the page for the data to appear.

driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
time.sleep(8)


# Scraping the images source.

ml=[]
for i in driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']"):
    ml.append(i.get_attribute('src'))
time.sleep(3)
    
# Keeping top 100 images.

ml= ml[0:100]
```

## Task 4

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
# Giving location to the driver.

driver.get('https://www.flipkart.com/')
```

```python
# Finding element for search bar and writing on it.

searchbar= driver.find_element_by_xpath("//input[@class='_3704LK']")
searchbar.send_keys('Pixel 4A')


# Finding element for search button and clicking on it

search_btn= driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
time.sleep(3)

# Scraping all the product urls.

urls=[]
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    urls.append(i.get_attribute('href'))
```

```python
# Creating lists to store data

brand=[]
phone=[]
color=[]
ram=[]
rom=[]
pc=[]
sc=[]
display=[]
resolution=[]
processor=[]
cores=[]
battery=[]
price=[]
```

```python
# Scraping all the required details, one product at a time.

for i in urls:
    driver.get(i)
    time.sleep(3)
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()     # Clicking on the read more option.
    try:
        b= driver.find_element_by_xpath("//span[@class='B_NuCI']")
        brand.append(b.text.split(' ')[0])
    except NoSuchElementException:
        brand.append('-')
    
    try:
        p= driver.find_element_by_xpath('//div[@class="_1UhVsV"]/div[1]/table/tbody/tr[3]/td[2]/ul/li')
        phone.append(p.text)
    except NoSuchElementException:
        phone.append('-')
        
    try:
        c= driver.find_element_by_xpath('//div[@class="_1UhVsV"]/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        color.append(c.text)
    except NoSuchElementException:
        color.append('-')
        
    try:
        r= driver.find_element_by_xpath('//div[@class="_1UhVsV"]/div[4]/table/tbody/tr[2]/td[2]/ul/li')
        ram.append(r.text)
    except NoSuchElementException:
        ram.append('-')
        
    try:
        o= driver.find_element_by_xpath('//div[@class="_1UhVsV"]/div[4]/table/tbody/tr[1]/td[2]/ul/li')
        rom.append(o.text)
    except NoSuchElementException:
        rom.append('-')
        
    try:
        m= driver.find_element_by_xpath('//div[@class="_2418kt"]/ul/li[3]')
        pc.append(m.text.split(' |')[0])
    except NoSuchElementException:
        pc.append('-')
        
    try:
        s= driver.find_element_by_xpath('//div[@class="_1UhVsV"]/div[5]/table/tbody/tr[5]/td[2]/ul/li')
        sc.append(s.text.replace('Yes','-'))
    except NoSuchElementException:
        sc.append('-')
        
    try:
        d= driver.find_element_by_xpath('//div[@class="_1UhVsV"]/div[2]/table/tbody/tr[1]/td[2]/ul/li')
        display.append(d.text)
    except NoSuchElementException:
        display.append('-')
        
    try:
        x= driver.find_element_by_xpath('//div[@class="_1UhVsV"]/div[2]/table/tbody/tr[2]/td[2]/ul/li')
        resolution.append(x.text.replace('$$',' '))
    except NoSuchElementException:
        resolution.append('-')
        
    try:
        e= driver.find_element_by_xpath("//div[@class='_2418kt']/ul/li[5]")
        processor.append(e.text)
    except NoSuchElementException:
        processor.append('-')
        
    try:
        k= driver.find_element_by_xpath("//li[contains(text(),'Core')]")
        cores.append(k.text)
    except NoSuchElementException:
        cores.append('-')
        
    try:
        t= driver.find_element_by_xpath('//div[@class="_2418kt"]/ul/li[4]')
        battery.append(t.text)
    except NoSuchElementException:
        battery.append('-')
        
    try:
        z= driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        price.append(z.text)
    except NoSuchElementException:
        price.append('-')
        
    time.sleep(3)
```

```python
# Creating dataframe

x= {'Brand':brand,'Phone': phone,'Colour': color, 'ROM': rom, 'RAM': ram, 'Primary camera': pc,'Secondary camera': sc,'Display size': display,
    'Resolution': resolution, 'Processor': processor,'Core':cores, 'Battery':battery,'Price':price,'Product URL': urls}
df3= pd.DataFrame(x)
df3
```

```python
# Saving the data as csv.

df3.to_csv('pixel4a_page.csv', index= False)
```

## Task 7

7. Write a program to scrap all the available details of top 10 gaming laptops from digit.in.

```python
# Giving address to the driver

driver.get('https://www.digit.in/')
```

```python
# Clicking through options to go the required page.

driver.find_element_by_xpath("//div[@class='menu']/ul/li[3]").click()  # clicking on the 'laptops' option.
time.sleep(4)
driver.find_element_by_xpath("//ul[@class='list-unstyled sidebar-list']//li[10]").click()  # clicking on the 'best gaming laptops' option.
```

```python
# Creating lists to store data.

name=[]
os=[]
display=[]
processor=[]
memory=[]
weight=[]
dimension=[]
graphic=[]
price=[]
```

```python
# Scraping all the details of the products.


for n in driver.find_elements_by_xpath("//div[@class='TopNumbeHeading sticky-footer' or @class='TopNumbeHeading sticky-footer active']/a/h3"):
    name.append(n.text)
    
for o in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[3]/td[3]"):
    os.append(o.text)
    
for d in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[4]/td[3]"):
    display.append(d.text)
    
for p in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[5]/td[3]"):
    processor.append(p.text)
    
for m in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[6]/td[3]"):
    memory.append(m.text)
    
for w in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[7]/td[3]"):
    weight.append(w.text)
    
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[8]/td[3]"):
    dimension.append(i.text)
    
for g in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[9]/td[3]"):
    graphic.append(g.text)
    
for p in driver.find_elements_by_xpath("//td[@class='smprice']"):
    price.append(p.text)
```

```python
# Creating dataframe

x= {'Name':name,'Operating system': os,'Display': display, 'Processor': processor, 'Memory': memory, 'Weight': weight,
    'Dimention': dimension,'Graphics processor': graphic, 'Price': price}
df6= pd.DataFrame(x)
df6
```
