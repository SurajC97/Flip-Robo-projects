# Task 1

1. Write a python program to display all the header tags from
   ‘en.wikipedia.org/wiki/Main_Page’.

```python
# sending request

page= requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

Response 200 means we can scrap data from this site.

```python
# source code

# page.content
```

```python
# making the content structured.

soup= BeautifulSoup(page.content)
#soup
```

```python
# making indentation.
    
soup.prettify()
```

```python
# extract the html tag where we have all the header tags.

header= soup.find_all('span', class_='mw-headline')
header
```

```python
header_list=[]
for i in header:
    header_list.append(i.text)
header_list
```

```python
header_list[1]= header_list[1].replace('\xa0','')
```

```python
print('Task 1: Displaying all headers:', '\n')
print(header_list)
```

2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name,

# Task 4

Write a python program to scrap book name, author name, genre and book review of
   any 5 books from ‘www.bookpage.com’

```python
page4= requests.get('https://bookpage.com/reviews')
page4
```

```python
soup4= BeautifulSoup(page4.content)
soup4.prettify()
```

```python
# Extracting book names

name= soup4.find_all('h4', class_='italic')
name
```

```python
names=[]
for i in name:
    names.append(i.text.replace('\n',''))
book_titles= names[0:5]
book_titles
```

```python
# Extracting book genre

genre= soup4.find_all('p', class_='genre-links hidden-phone')
genre
```

```python
genres=[]
for i in genre:
    genres.append(i.text.replace('\n',''))
book_genre= genres[0:5]
book_genre
```

```python
# Extracting book author names

author= soup4.find_all('p', class_='sans bold')
author
```

```python
authors=[]
for i in author:
    authors.append(i.text.replace('\n',''))
book_author= authors[0:5]
book_author
```

```python
# Extracting book reviews

rev= soup4.find_all('p', class_='excerpt')
rev
```

```python
reviews=[]
for i in rev:
    reviews.append(i.text.replace('\n',''))
book_review= reviews[0:5]
book_review
```

We can see 1 review is missing. They do not belong to the tag used having class name 'excerpt'.

```python
# lets check all the p tags inside div tag.

z=[]
div= soup4.div
for i in div.find_all('p'):
    z.append(i.text.replace('\n',''))
    
z
```

Now we extract the required reviews of our books.

```python
frst_rev=[z[3],z[7]]
frst_rev
```

```python
last_rev=book_review[2:5]
last_rev
```

```python
# joining our first two and last three reviews.

all_book_review=frst_rev+last_rev
all_book_review              
```

```python
# Creating a data frame of the book data.

import pandas as pd
pd.set_option('display.max_colwidth',None)
book= {'Title':book_titles, 'Author': book_author, 'Genre': book_genre, 'Review': all_book_review}
df3= pd.DataFrame(book)
df3
```
