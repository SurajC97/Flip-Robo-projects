Q1. Scrape the details of most viewed videos on YouTube from Wikipedia: Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/ You need to find following details: A) Rank B) Name C) Artist D) Upload date E) Views

```python
# Giving address to the driver

driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
# Creating lists to store data

rank=[]
name=[]
artist=[]
date=[]
views=[]
```

```python
# Scraping data

for i in range(1,31):
    
    # Scraping rank
    r = driver.find_elements_by_xpath('//div[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[{}]/td[1]'.format(i))
    for j in r:
        rank.append(j.text)
        
    # Scraping name
    n = driver.find_elements_by_xpath('//div[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[{}]/td[2]'.format(i))
    for j in n:
        name.append(j.text.split('[')[0])
        
    # Scraping artist
    a = driver.find_elements_by_xpath('//div[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[{}]/td[3]'.format(i))
    for j in a:
        artist.append(j.text)
        
    # Scraping views
    v = driver.find_elements_by_xpath('//div[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[{}]/td[4]'.format(i))
    for j in v:
        views.append(j.text + ' Billions')
        
    # Scraping date
    d = driver.find_elements_by_xpath('//div[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[{}]/td[5]'.format(i))
    for j in d:
        date.append(j.text)
        
```

```python
# Creating dataframe

x= {'Ranking':rank, 'Name': name, 'Artist': artist, 'Views': views, 'Upload Date': date}
df1= pd.DataFrame(x)
df1
```

Q2. Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/. You need to find following details: A) Match title (I.e. 1st ODI) B) Series C) Place D) Date E) Time Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# opening required page on the browser.

driver.get('https://www.bcci.tv/')
```

```python
# Extracting the elements of the 'international' button and the 'fixtures' button which is inside hover and clicking on it.

intr= driver.find_element_by_xpath("//ul[@class='navigation__list showMoreEnabled']/li[1]/div[1]")
fix= driver.find_element_by_xpath("//ul[@class='navigation__list showMoreEnabled']/li[1]/div[2]/div/ul/li[1]")
action= ActionChains(driver)
action.move_to_element(intr).move_to_element(fix).click().perform()
```

```python
# Creating lists to store data

title=[]
series=[]
place=[]
day=[]
month=[]
date=[]
time=[]

# Scraping the data

t = driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
for j in t:
    title.append(j.text)
    
s = driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]')
for j in s:
    series.append(j.text)
    
p = driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span')
for j in p:
    place.append(j.text)
        
d = driver.find_elements_by_xpath('//span[@class="fixture__day"]')
for j in d:
    day.append(j.text)
    
dt = driver.find_elements_by_xpath('//span[@class="fixture__date"]')
for j in dt:
    date.append(j.text)
        
m = driver.find_elements_by_xpath('//span[@class="fixture__month"]')
for j in m:
    month.append(j.text)        
                
i = driver.find_elements_by_xpath('//span[@class="fixture__time"]')
for j in i:
    time.append(j.text)
```

```python
# Creating dataframe

x= {'Title':title, 'Series': series, 'Place': place, 'Day': day, 'Date': date, 'Month': month, 'Time':time}
df2= pd.DataFrame(x)

# Combining the date and months

df2['Fixture Date']=df2['Date'].astype(str)+' '+df2['Month']+' '+df2['Day']
df2.drop(['Day','Month','Date'], axis=1, inplace=True)
df2
```

Q4. Scrape the details of State-wise GDP of India from statisticstime.com. Url = http://statisticstimes.com/ You have to find following details: A) Rank B) State C) GSDP at current price (19-20) D) GSDP at current price (18-19) E) Share(18-19) F) GDP($ billion) Note: - From statisticstimes home page you have to reach to economy page through code.

```python
# opening required page on the browser.

driver.get('http://statisticstimes.com/')
```

```python
# Clicking on option 'India', which is hovering inside the 'Economy' option.

economy= driver.find_element_by_xpath("//div[@class='navbar']/div[2]/button")
india= driver.find_element_by_xpath("//div[@class='navbar']/div[2]/div/a[3]")

action= ActionChains(driver)
action.move_to_element(economy).move_to_element(india).click().perform()
time.sleep(3)

# Clicking on GDP of Indian states.

driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/ul/li[1]/a').click()
```

```python
# Creating lists to store data

rank=[]
state=[]
crp19=[]
crp18=[]
share=[]
gdp=[]

# Scraping the data

for i in range(1,34):
    
    # Scraping rank
    r = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr[{}]/td[1]'.format(i))
    for j in r:
        rank.append(j.text)
        
    # Scraping state
    s = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr[{}]/td[2]'.format(i))
    for j in s:
        state.append(j.text)
        
    # Scraping GSDP
    c = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr[{}]/td[3]'.format(i))
    for j in c:
        crp19.append(j.text)
        
    # Scraping GSDP
    p = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr[{}]/td[4]'.format(i))
    for j in p:
        crp18.append(j.text)
        
    # Scraping share
    h = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr[{}]/td[5]'.format(i))
    for j in h:
        share.append(j.text)
        
    # Scraping GDP
    g = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr[{}]/td[6]'.format(i))
    for j in g:
        gdp.append(j.text)
```

```python
# Creating dataframe

x= {'Rank':rank, 'State': state, 'GSDP at current price(19-20)': crp19, 'GSDP at current price(18-19)':crp18, 'Share': share, 'GDP($billion)': gdp}
df4= pd.DataFrame(x)
df4
```

Q5. Scrape the details of trending repositories on Github.com. Url = https://github.com/ You have to find the following details: A) Repository title B) Repository description C) Contributors count D) Language used ASSIGNMENT Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# opening required page on the browser.

driver.get('https://github.com/')
```

```python
# Opening the trending page through the explore menu.

try:
    explore= driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]/details/summary")
    trending= driver.find_element_by_xpath("//div[@class='dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4']/ul[2]/li[3]/a")
    action= ActionChains(driver)
    action.move_to_element(explore).move_to_element(trending).click().perform()
except ElementNotInteractableException:
    driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]/details/summary").click()
    time.sleep(2)
    driver.find_element_by_xpath("//div[@class='dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4']/ul[2]/li[3]/a").click()
```

```python
# Creating lists to store data

name=[]
desc=[]
count=[]
lang=[]

# Scraping the repository title
name=[]
n= driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in n:
    name.append(i.text)
    

# Scraping the repository description
for i in range(1,26):
    try:
        d= driver.find_element_by_xpath("//div[@class='Box']/div[2]/article[{}]/p".format(i))
        desc.append(d.text)
    except NoSuchElementException:
        desc.append('-')
        
        
# Scraping the contributors count
c= driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']/a[2]")
for i in c:
    count.append(i.text)
    
    
# Scraping the language used
for i in range(1,26):
    try:
        l= driver.find_element_by_xpath("//div[@class='Box']/div[2]/article[{}]/div[2]/span[1]/span[2]".format(i))
        lang.append(l.text)
    except NoSuchElementException:
        lang.append('-')
```

```python
# Creating dataframe

x= {'Repository Title':name, 'Repository Description': desc, 'Contributors Count': count, 'Language used':lang}
df5= pd.DataFrame(x)
df5
```

```python
# Creating lists to store data

name=[]
des=[]
com=[]
skill=[]
loc=[]


# Scraping name
n = driver.find_elements_by_xpath('//span[@class="fl ellipsis"]')
for i in n:
    name.append(i.text)
    
# Scraping designation
d = driver.find_elements_by_xpath('//span[@class="ellipsis clr"]')
for i in d:
    des.append(i.text)

# Scraping company name 
c = driver.find_elements_by_xpath('//div[@class="vcard"]/p/a[2]')
for i in c:
    com.append(i.text)
    
# Scraping the skills they hire for 
s = driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]')
for i in s:
    skill.append(i.text)
    
# Scraping the location 
for i in range(1,26):
    for j in range(1,3):
        try:
            l= driver.find_element_by_xpath('//*[@id="tabP-1"]/div[{}]/div[{}]/div[1]/div[1]/p/span[2]/small'.format(i,j))
            loc.append(l.text)
        except NoSuchElementException:
            loc.append('-')
```

```python
# Creating lists to store data

book=[]
author=[]
sold=[]
publisher=[]
genre=[]


# Scraping data

for i in range(0,100):
    
    # Scraping the book name
    
    b = driver.find_elements_by_id("table-cell-10943-{}-1".format(i))
    for j in b:
        book.append(j.text)
     
    
    # Scraping the author name
    
    a = driver.find_elements_by_id("table-cell-10943-{}-2".format(i))
    for j in a:
        author.append(j.text)
      
    
    # Scraping the volumes sold
    
    s = driver.find_elements_by_id("table-cell-10943-{}-3".format(i))
    for j in s:
        sold.append(j.text)
        
        
    # Scraping the publisher
    
    p = driver.find_elements_by_id("table-cell-10943-{}-4".format(i))
    for j in p:
        publisher.append(j.text)
        
        
    # Scraping the genre
    
    g = driver.find_elements_by_id("table-cell-10943-{}-5".format(i))
    for j in g:
        genre.append(j.text)
```

```python
# Creating dataframe

x= {'Book Name':book, 'Author Name': author, 'Volumes Sold': sold, 'Publisher':publisher, 'Genre': genre}
df8= pd.DataFrame(x)
df8
```

Q10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details: A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year Note: - from the home page you have to go to the Show All Dataset page through code.

```python
# opening required page on the browser.

driver.get('https://archive.ics.uci.edu/')
```

```python
# Clicking on 'View all dataset'

driver.find_element_by_xpath("//b[contains(text(),'View ALL Data Sets')]").click()
```

```python
# Creating lists to store data

name=[]
dtype=[]
task=[]
atype=[]
instn=[]
nattri=[]
year=[]


# Scraping the data

n = driver.find_elements_by_xpath('//p[@class="normal"]/b/a')
for i in n:
    name.append(i.text)
    
for i in range(2,590):
    d= driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[2]/p".format(i))
    dtype.append(d.text)
    
    t= driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[3]/p".format(i))
    task.append(t.text)
    
    a= driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[4]/p".format(i))
    atype.append(a.text)
    
    s= driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[5]/p".format(i))
    instn.append(s.text)
    
    n= driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[6]/p".format(i))
    nattri.append(n.text)
    
    y= driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[7]/p".format(i))
    year.append(y.text)
```

```python
# Creating dataframe

x= {'Dataset Name':name, 'Data Type': dtype, 'Task': task, 'Attribute Type': atype, 'No of Instances': instn, 'No of Attribute':nattri, 'Year': year}
df10= pd.DataFrame(x)
df10
```
